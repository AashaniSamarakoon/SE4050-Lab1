```python
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5, 5, 201)
y = np.linspace(-5, 5, 201)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_xlabel('X'); ax.set_ylabel('Y'); ax.set_zlabel('Z')
ax.set_title('Z = X^2 + Y^2')
fig.colorbar(surf, shrink=0.6, aspect=10)
plt.show()
```

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv

df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

candidates = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed',
              'Sp. Ark', 'SpAtk', 'Sp Atk', 'Sp_Atk']
found = [c for c in df.columns if c in candidates]
print("Using columns:", found)

pearson = df[found].corr(method='pearson')
spearman = df[found].corr(method='spearman')

plt.figure(figsize=(6,5))
sns.heatmap(pearson, annot=True, fmt=".2f", vmin=-1, vmax=1, cmap='coolwarm')
plt.title("Pearson correlation")
plt.show()

plt.figure(figsize=(6,5))
sns.heatmap(spearman, annot=True, fmt=".2f", vmin=-1, vmax=1, cmap='coolwarm')
plt.title("Spearman correlation")
plt.show()
```
